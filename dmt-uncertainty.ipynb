# Discrete Morse Theory

```python
from dmt_2d import DMT
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import IPython
```

```python
np_path = "./data.npy"
np_pred = np.load(np_path)
print(np_pred.shape)
```

```python
dmtclass = DMT(np_pred)
```

```python
img_likelihood = dmtclass.showLikelihood()
plt.title("Likelihood map (output of UNet)")
plt.imshow(img_likelihood, cmap='gray')
```

```python
img_binary = img_likelihood.copy()
img_binary[img_binary >= 128] = 255
img_binary[img_binary < 128] = 0
plt.title("Corresponding binary segmentation")
plt.imshow(img_binary, cmap='gray')
```

```python
img_crit = dmtclass.showCriticalPoints()
plt.title("Critical points (red = saddle; blue = maxima)")
plt.imshow(img_crit)
```

```python
img_pieces = dmtclass.showManifolds()
plt.title("Manifold pieces")
plt.imshow(img_pieces)
```

```python
img_skelt = dmtclass.showSkeleton()
plt.title("Morse skeleton (recovered false negatives)")
plt.imshow(img_skelt)
```

```python
img_gt = np.array(Image.open("./sample_C60_along_6edge_neg_pol_training_00_0_0_edge.png"))
plt.title("Ground truth")
plt.imshow(img_gt, cmap='gray')
```

```python
fig, ax = plt.subplots(ncols=3, nrows=2)
ax[0,0].set_title("Likelihood pred")
ax[0,0].imshow(img_likelihood, cmap='gray')
ax[0,1].set_title("Critical points")
ax[0,1].imshow(img_crit)
ax[0,2].set_title("Manifold pieces")
ax[0,2].imshow(img_pieces)
ax[1,0].set_title("Binary pred")
ax[1,0].imshow(img_binary, cmap='gray')
ax[1,1].set_title("Morse skeleton")
ax[1,1].imshow(img_skelt)
ax[1,2].set_title("Ground truth")
ax[1,2].imshow(img_gt, cmap='gray')
```
